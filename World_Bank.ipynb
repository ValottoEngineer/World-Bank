```python
import pandas as pd
import matplotlib.pyplot as plt
from pandas_datareader import wb

indicator = 'NY.GDP.PCAP.CD'
countries = ['all']
start_year = 2019
end_year = 2019

data = wb.download(indicator=indicator, country=countries, start=start_year, end=end_year)

# Resetar o índice para facilitar o trabalho com Pandas
data.reset_index(inplace=True)

data.columns = ['Country', 'Year', 'GDP per capita (current US$)']

# Remover linhas com valores nulos, se houver
data.dropna(inplace=True)

# Visualizar as primeiras linhas do DataFrame para verificar se os dados foram carregados corretamente
print(data.head())

```

**Agora, vamos ver como fica com os gráficos solicitados:**

# Gráfico de Linha

```python
plt.figure(figsize=(10, 6))
plt.plot(data['Country'], data['GDP per capita (current US$)'], marker='o', linestyle='-', color='b')
plt.title('GDP per capita (current US$) - 2019')
plt.xlabel('Country')
plt.ylabel('GDP per capita (current US$)')
plt.xticks(rotation=90)
plt.grid(True)
plt.tight_layout()
plt.show()
```

# Gráfico de Barra


```python
plt.figure(figsize=(12, 8))
plt.bar(data['Country'], data['GDP per capita (current US$)'], color='skyblue')
plt.title('GDP per capita (current US$) - 2019')
plt.xlabel('Country')
plt.ylabel('GDP per capita (current US$)')
plt.xticks(rotation=90)
plt.grid(True)
plt.tight_layout()
plt.show()
```

# Histograma

```python
plt.figure(figsize=(10, 6))
plt.hist(data['GDP per capita (current US$)'], bins=20, color='orange', edgecolor='black')
plt.title('Histograma de GDP per capita (current US$) - 2019')
plt.xlabel('GDP per capita (current US$)')
plt.ylabel('Frequência')
plt.grid(True)
plt.tight_layout()
plt.show()
```

# Gráfico de Pizza

```python
top_countries = data.nlargest(10, 'GDP per capita (current US$)')
plt.figure(figsize=(8, 8))
plt.pie(top_countries['GDP per capita (current US$)'], labels=top_countries['Country'], autopct='%1.1f%%', startangle=140)
plt.title('Top 10 Países por GDP per capita (current US$) - 2019')
plt.axis('equal')
plt.tight_layout()
plt.show()
```

# Gráfico de Scatter (dispersão)

```python
import numpy as np

np.random.seed(0)
data['Life Expectancy'] = np.random.uniform(50, 85, size=len(data))

plt.figure(figsize=(10, 8))
plt.scatter(data['GDP per capita (current US$)'], data['Life Expectancy'], alpha=0.7)
plt.title('Scatter Plot: GDP per capita vs Expectativa de Vida - 2019')
plt.xlabel('GDP per capita (current US$)')
plt.ylabel('Expectativa de Vida')
plt.grid(True)
plt.tight_layout()
plt.show()
```
